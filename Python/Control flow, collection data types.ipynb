**Arithmetic operators**

```
# Arithmetic operators ----------------------------

x = 15
y = 4

# Output: x + y = 19
print('x + y =',x+y)
# Output: x - y = 11
print('x - y =',x-y)
# Output: x * y = 60
print('x * y =',x*y)
# Output: x / y = 3.75
print('x / y =',x/y)
# Output: x // y = 3
print('x // y =',x//y)
# Output: x ** y = 50625
print('x ** y =',x**y)
```

**Comparision operators**

```
# comparision operators ----------------------------

x = 10
y = 12

# Output: x > y is False
print('x > y  is',x>y)
# Output: x < y is True
print('x < y  is',x<y)
# Output: x == y is False
print('x == y is',x==y)
# Output: x != y is True
print('x != y is',x!=y)
# Output: x >= y is False
print('x >= y is',x>=y)
# Output: x <= y is True
print('x <= y is',x<=y)
```

## Collection data types

* List is a collection which is ordered and changeable. Allows duplicate members.
* Tuple is a collection which is ordered and unchangeable. Allows duplicate members.
* Set is a collection which is unordered and unindexed. No duplicate members.
* Dictionary is a collection which is unordered, changeable and indexed. No duplicate members.


* Use a tuple when you know what information goes in the container that it is. For example, when you want to store a person’s credentials for your website. think of it as a row in a DB table.  It is read-only

* But when you want to store similar elements, like in an array in C++, you should use a list.  Think of it as a collection of similar values.

* In practical applications, lists are very nice to sort and have order while sets are nice to use when you don't want duplicates and don't care about order.  You can use set operations on sets.

* Use a dictionary if the purpose is to store key value pairs like configuration settings.

**Sets**

* difference() - Returns a set containing the difference between two or more sets
* intersection() - Returns a set, that is the intersection of two other sets
* isdisjoint() - Returns True if the intersection is null
* issubset() - Returns whether another set contains this set or not
* issuperset() - Returns whether this set contains another set or not
* union() - Return a set containing the union of sets


```
## Sets--------------------

thisset = {1,2,3}
print(thisset)

thisset = {1,2,3}

for x in thisset:
  print(x)

thisset = {1,2,3}
print(4 in thisset)

thisset = {1,2,3}
thisset.add(4)
print(thisset)

thisset = {1,2,3}
thisset.update([4,5,6])
print(thisset)

thisset = {1,2,3}
print(len(thisset))

thisset = {1,2,3}
thisset.remove(2)
print(thisset)


thisset = {1,2,3}
thisset.clear()
print(thisset)

thisset = {1,2,3}
del thisset
#print(thisset)

# initialize A and B
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}
C = {1,2}

print(A.union(B))
print(A.intersection(B))
print(A.isdisjoint(C))
print(A.issuperset(C))
print(A.issubset(C))
print(A.difference(B))
print(B.difference(A))
```

**Dictionaries**

* items() - Returns a list containing the a tuple for each key value pair
* keys() - Returns a list containing the dictionary's keys
* values() - Returns a list of all the values in the dictionary


```
## Dictionaries--------------------

thisdict =	{
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
print(thisdict)
x = thisdict["model"]
print(x)

thisdict =	{
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
thisdict["year"] = 2018

for x in thisdict:
  print(x)

for x in thisdict.keys():
  print(x)

for x in thisdict.values():
  print(x)

for x, y in thisdict.items():
  print(x, y)

print(len(thisdict))

thisdict =	{
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
thisdict["color"] = "red" # add new one
print(thisdict)

thisdict =	{
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
thisdict.pop("model")
print(thisdict)

thisdict =	{
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
del thisdict
#print(thisdict) #this will cause an error because "thislist" no longer exists.

thisdict =	{
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
thisdict.clear()
print(thisdict)
```
